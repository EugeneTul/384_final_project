```python
import pandas as pd
import os
from glob import glob
```

```python
#November load
folder_path = '/Users/eugenetulyagijja/Documents/Eugene Tulyagijja/school/Fall 2023/SAL 384/Final Project'
```

```python
excel_files = glob(os.path.join(folder_path, '*.xlsx'))
```

```python
dfs = []
```

```python
for file in excel_files:
    df = pd.read_excel(file)
    dfs.append(df)
```

```python
final_df = pd.concat(dfs, ignore_index=True)
final_df
```

```python
#Creating a mac from distance which is just (max speed - Distnace/100) Formula is provided from Kinexon

final_df['Max Cumal'] = (final_df['Speed (max.) (mph)'] - final_df['Distance (mi)']) / 100
```

```python
final_df
```

```python
#multiplying mecchanical load to mac cumal 
final_df['Max Cumal * Mechanical Load'] = final_df['Max Cumal'] * final_df['Mechanical Load']
final_df
```
